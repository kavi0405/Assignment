##### Q2: This assignment is for visualization using matplotlib: data to use: url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

##### Create a pie chart presenting the male/female proportion

```python
labels=['male','female']
sizes=titanic.sex.value_counts()
```

```python
fig1,ax1=plt.subplots()
ax1.pie(sizes, labels=labels, startangle=90, autopct='%.1f%%', shadow=True, colors=['Blue','Red'])
plt.title('Male and Female Proportion: Titanic data')
plt.show()
```

##### Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure()
category=titanic[titanic.sex=='male'].plot.scatter('age','fare',color='blue',label='MALE')
titanic[titanic.sex=='female'].plot.scatter('age','fare',color='red',label='FEMALE', ax=category)
plt.title('Scatterplot with Fare paid, Age by Gender: Titanic')
```
