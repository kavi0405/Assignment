### Assignment 8: Data Visualization

##### Q: transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
# import iris dataset
iris=load_iris()
```

```python
# Transforming the dataset to have 3 dimensions using PCA for feature reduction
pca = PCA(n_components=3)
pca_iris = pca.fit_transform(iris['data'])
```

```python
# visualizing the transformed data in 3D form
fig = plt.figure(1, figsize=(4, 3))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(pca_iris[:,0], pca_iris[:,1], pca_iris[:,2], c=iris['target'], cmap=plt.cm.Set1, edgecolor='k', s=40)

ax.set_title("Iris Data after transformation")

plt.show()
```
