### Assignment 23: ML Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy import misc
```

```python
# Load the dataset

face = misc.face(gray=True)
```

```python
X = face.reshape((-1,1))
```

```python
X
```

```python
# KMeans clustering

km = KMeans(n_clusters=5, n_init=10)
km.fit(X)
```

```python
values = km.cluster_centers_.squeeze()
labels = km.labels_
```

```python
values = values.astype('int')
values
```

```python
# Compression

face_comp = np.choose(labels,values)
face_comp.shape = face.shape
```

```python
vmin=face.min()
vmax=face.max()
```

```python
plt.imshow(face, cmap=plt.cm.gray,vmin=vmin, vmax=256)
plt.title('Original Image')
```

```python
plt.imshow(face_comp, cmap=plt.cm.gray, vmin=vmin, vmax=vmax)
print('Compressed Image')
```
